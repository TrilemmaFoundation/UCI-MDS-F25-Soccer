```python
# # Choose a match (example: Russia vs Denmark)
# match_info = pd.read_sql_query("""
# SELECT match_id, home_team, away_team, home_score, away_score, match_date
# FROM matches
# WHERE home_team = 'Russia' AND away_team = 'Denmark';
# """, conn)

# print("Match info:")
# print(match_info)

# # Extract the match_id
# match_id = int(match_info.iloc[0]['match_id'])
# print(f"\nSelected match_id: {match_id}")

# # Load all events for this match
# events = pd.read_sql_query(f"""
# SELECT type, team, location_x, location_y, minute, second
# FROM events
# WHERE match_id = {match_id}
#   AND location_x IS NOT NULL
#   AND location_y IS NOT NULL;
# """, conn)

# print("\nEvent sample:")
# events.head()




# Load all events for this match (Russia vs Denmark)
events = pd.read_sql_query(f"""
SELECT type, team, location, pass_end_location, shot_end_location, carry_end_location
FROM events
WHERE match_id = 3788769;
""", conn)

# Helper to parse "[x, y]" or JSON-like strings
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def parse_loc(value):
    if value is None:
        return None, None
    try:
        loc = json.loads(value)
        if isinstance(loc, list) and len(loc) >= 2:
            return float(loc[0]), float(loc[1])
        else:
            return None, None
    except:
        try:
            vals = value.strip('[]').split(',')
            return float(vals[0]), float(vals[1])
        except:
            return None, None

# Create unified start position from `location`
events['location_x'], events['location_y'] = zip(*events['location'].map(parse_loc))
events = events.dropna(subset=['location_x', 'location_y']).copy()

print("Teams found:", events["team"].unique())
events.head(5)


```

```python
# Replace with the exact names from previous print
team_a = "Russia"
team_b = "Denmark"

# Filter events for both teams
passes_a = events[(events['type'] == 'Pass') & (events['team'] == team_a)]
passes_b = events[(events['type'] == 'Pass') & (events['team'] == team_b)]

shots_a = events[(events['type'] == 'Shot') & (events['team'] == team_a)]
shots_b = events[(events['type'] == 'Shot') & (events['team'] == team_b)]

receipts_a = events[(events['type'] == 'Ball Receipt*') & (events['team'] == team_a)]
receipts_b = events[(events['type'] == 'Ball Receipt*') & (events['team'] == team_b)]

print(f"{team_a}: {len(passes_a)} passes, {len(shots_a)} shots, {len(receipts_a)} receipts")
print(f"{team_b}: {len(passes_b)} passes, {len(shots_b)} shots, {len(receipts_b)} receipts")
```

```python
def side_by_side_heatmap(x1, y1, x2, y2, team1, team2, title, cmap1='turbo', cmap2='turbo'):
    """
    Generate side-by-side 2D histograms for two teams.
    """
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    for ax, x, y, team, cmap in zip(axes, [x1, x2], [y1, y2], [team1, team2], [cmap1, cmap2]):
        heatmap, xedges, yedges = np.histogram2d(x, y, range=[[0,120],[0,80]], bins=50)
        extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
        im = ax.imshow(heatmap.T, extent=extent, origin='lower', cmap=cmap)
        ax.set_xlim(0,120)
        ax.set_ylim(0,80)
        ax.set_xlabel("Pitch X (0–120)")
        ax.set_ylabel("Pitch Y (0–80)")
        ax.set_title(f"{team}", fontsize=14)
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    plt.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

side_by_side_heatmap(
    passes_a['location_x'], passes_a['location_y'],
    passes_b['location_x'], passes_b['location_y'],
    team_a, team_b, title="Pass Locations — Russia vs Denmark"
)

side_by_side_heatmap(
    shots_a['location_x'], shots_a['location_y'],
    shots_b['location_x'], shots_b['location_y'],
    team_a, team_b, title="Shot Locations — Russia vs Denmark"
)

side_by_side_heatmap(
    receipts_a['location_x'], receipts_a['location_y'],
    receipts_b['location_x'], receipts_b['location_y'],
    team_a, team_b, title="Ball Receipts — Russia vs Denmark"
)
```

```python
match_id = 3788769

# Query passes by 5-minute intervals
query = f"""
WITH pass_intervals AS (
    SELECT
        team,
        ((minute / 5) * 5) AS interval_start,
        COUNT(*) AS pass_count
    FROM passes
    WHERE match_id = {match_id}
    GROUP BY team, interval_start
)
SELECT * FROM pass_intervals
ORDER BY interval_start, team;
"""

df = pd.read_sql_query(query, conn)
conn.close()

# Pivot for side-by-side team comparison
pivot_df = df.pivot(index="interval_start", columns="team", values="pass_count").fillna(0)

# Plot line graph
plt.figure(figsize=(10, 6))
for team in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[team], marker='o', label=team)

plt.title("Passes per 5-Minute Interval by Team")
plt.xlabel("Time Interval (minutes)")
plt.ylabel("Number of Passes")
plt.xticks(range(0, int(df["interval_start"].max() + 5), 5))
plt.legend(title="Team")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
